## **Training the first machine learning model for this data**

```python
"""
Creates a new column "target" and sets it equal to the temp_max from the next day (row) -> because we're trying to predict the next day's 
temp_max, so the target would be the next day's actual temp_max value
"""
core_data["target"] = core_data.shift(-1)["temp_max"]
```

```python
core_data
```

```python
# Deleting the last row because the target value is null (it's null since we don't have data for 03/10)
core_data = core_data.iloc[:-1, :].copy()
```

```python
core_data
```

```python
from sklearn.linear_model import Ridge # Ridge regression is a type of regression that minimizes overfitting by using a penalty term (alpha)

regression = Ridge(alpha=0.1)
```

```python
predictors = ["precipitation", "snow", "snow_depth", "temp_max", "temp_min"]
```

```python
training = core_data.loc[:"2022-12-31", :] # training data
```

```python
testing = core_data.loc["2023-01-01":, :] # testing data
```

```python
regression.fit(training[predictors], training["target"]) # Using predictors to train the training data to try to predict "target"
```

```python
predictions = regression.predict(testing[predictors]) # Generate predictions on test data set using the predictor columns
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(testing["target"], predictions)
```

```python
# Combining actual data with predictions -> this will help me see the differences between the actual data and predictions
# To do that, I converted predictions to a series
combined_data = pd.concat([testing["target"], pd.Series(predictions, index = testing.index)], axis = 1)
combined_data.columns = ["actual", "predictions"]
```

```python
combined_data
```

```python
combined_data.plot()
```

```python
regression.coef_
```

```python
# Created a function to do what I did in the previous lines of code
def create_predictions(predictors, core_data, regression):
    training = core_data.loc[:"2022-12-31", :] # training data
    testing = core_data.loc["2023-01-01":, :] # testing data
    regression.fit(training[predictors], training["target"]) # Using predictors to train the training data to try to predict "target"
    predictions = regression.predict(testing[predictors]) # Generate predictions on test data set using the predictor columns
    error = mean_absolute_error(testing["target"], predictions)
    
    # Combining actual data with predictions -> this will help me see the differences between the actual data and predictions
    # To do that, I converted predictions to a series
    combined_data = pd.concat([testing["target"], pd.Series(predictions, index = testing.index)], axis = 1)
    combined_data.columns = ["actual", "predictions"] # Renaming the columns

    return error, combined_data
```

```python
# Finds the average temp_max value in the past 30 days for every day and assigns that value to the "month_max" column in core_data
core_data["month_max"] = core_data["temp_max"].rolling(30).mean()
```

```python
core_data
```

```python
core_data = core_data.iloc[30:, :].copy()
```

```python
core_data
```

```python
# Finding how much the average temp_max value in the past 30 days for every day varies from the given temp_max for that day
# I used the approach below to avoid getting division by zero errors
core_data["month_day_max"] = np.where(core_data["temp_max"] != 0, core_data["month_max"] / core_data["temp_max"], np.nan)
```

```python
# Ratio of temp_max to temp_min
core_data["max_min"] = np.where(core_data["temp_min"] != 0, core_data["temp_max"] / core_data["temp_min"], np.nan)
```

```python
month_day_max_nan = core_data["month_day_max"].isna().sum()
max_min_nan = core_data["max_min"].isna().sum()

print("Number of nan values in month_day_max column:", month_day_max_nan)
print("Number of nan values in max_min_nan column:", max_min_nan)
```

```python
# Finding for which dates these nan values occur in the max_min column
nan_rows = core_data[core_data["max_min"].isna()]
nan_rows[["max_min"]]
```

We can see that the NaN values are present in the max_min column for dates that will be used for our training set data 
(refer to create_predictions function for the training data)

I will now find the percentage of rows in the max_min column in the training data set that would contain nan values. 

```python
# Finding the percentage of rows in the max_min column in the training data set that would contain nan values. 
training_data = core_data.loc[:"2022-12-31", :]
num_rows = len(training_data)
percent_rows_nan = (max_min_nan / num_rows) * 100

print("Percentage of training data that contains nan values:", percent_rows_nan)
```

```python
# Since a very small percentage of the training data contains nan values in the max_min column, I'm dropping those rows
# Also, we get rid of nan values because ridge regression does not accept nan values as inputs
core_data = core_data.dropna(subset=["max_min"]).copy()
```

```python
# Add these newly created columns to the list of predictor columns
predictors = ["precipitation", "snow", "snow_depth", "temp_max", "temp_min", "month_max", "month_day_max", "max_min"]
```

```python
error, combined_data = create_predictions(predictors, core_data, regression)
```

```python
error
```

```python
combined_data.plot()
```

```python
# Finds the historical monthly average temperature for each month

# Resets the index of core_data, sets it equal to a separate dataframe that we can use
core_data_reset = core_data.reset_index() 

# Calculate expanding mean values
expanding_mean = core_data_reset.groupby(core_data_reset["DATE"].dt.month)["temp_max"].expanding(1).mean().reset_index(level=0, drop=True)

# Set the index of expanding_mean equal to core_data so that there are no issues
expanding_mean.index = core_data.index

# Assign expanding mean values to a new column in core_data
core_data["monthly_average"] = expanding_mean
```

```python
core_data
```

```python
# Finding the average of all the previous xth days of the month to compute the day_of_year_average
    # For example, the "day_of_year_average" for the 30th day of the month is = to the average of all the 30th days of the previous months 

# Resets the index of core_data, sets it equal to a separate dataframe that we can use
core_data_reset = core_data.reset_index() 

# Calculate expanding mean values
expanding_mean = core_data_reset.groupby(core_data_reset["DATE"].dt.dayofyear)["temp_max"].expanding(1).mean().reset_index(level=0, drop=True)

# Set the index of expanding_mean equal to core_data so that there are no issues
expanding_mean.index = core_data.index

core_data["day_of_year_average"] = expanding_mean
```

```python
core_data
```

```python
# Let's update the predictors list to add the new predictors I created (monthly_average and day_of_year_average)
predictors = ["precipitation", "snow", "snow_depth", "temp_max", "temp_min", "month_max", "month_day_max", "max_min", "monthly_average", "day_of_year_average"]
```

```python
error, combined_data = create_predictions(predictors, core_data, regression)
```
